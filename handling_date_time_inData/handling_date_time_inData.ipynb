```python
# Handling date and time in the data implementation
# The code having two datasets as orders and messages
# The orders will handle the date while messages is for the time

# Author: Muhammad Humayun Khan

import numpy as np
import pandas as pd

dataset_date = '/content/drive/MyDrive/datasets/orders.csv'     # date dataset
dataset_time = '/content/drive/MyDrive/datasets/messages.csv'   # time dataset

date = pd.read_csv(dataset_date)
time = pd.read_csv(dataset_time)

date.head()
```

**Working with date**

```python
# Converting to datetime datatype
# by default setting in dataset for the date and time is string so better to convert to date and time type
# in order to work better on it
date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
# Extract year - a new column will be added at the end
date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
# Extract month - new column added at the end
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
# lets add name of the month - new column added at the end
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
# extract days
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract quarter
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract semester
# 1 = August/September to December
# 2 - January to May.
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract time elapsed between date
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
